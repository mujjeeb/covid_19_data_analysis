```python
import sqlalchemy
import psycopg2
from sqlalchemy import create_engine
import pandas as pd
```

```python
#Reading CSV file from Google docs
sheet_url = 'https://docs.google.com/spreadsheets/d/1EyM7scZHltBXZtZIjgOovarFdfdFv-LBWVYkGEi7sts/edit#gid=53936285'
data_url = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')

data = pd.read_csv(data_url,index_col=False)
```

```python
data
```

```python
#Remove index from data as index is not a columnin the Table created on PGadmin
df = data.set_index('SNo')
```

```python
df
```

```python
# Connecting the engine to the pre-established table
engine = create_engine('postgresql://postgres:[Password]@localhost:5432/covid_19_data')
df.to_sql('covid_19_data', engine, if_exists='append')
```
